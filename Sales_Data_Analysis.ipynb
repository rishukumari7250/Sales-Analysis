***Import Libraries***

```python
import pandas as pd
import matplotlib.pyplot as plt
```

***Load Data***

```python
df = pd.read_csv('/content/train.csv')
df.head()
```

***Basic Data Cleaning***

```python
df . info()
df. isnull().sum()
```

### *Data Analysis*

***Total Sales***

```python
total_sales = df['Sales'].sum()
print("Total Sales:",total_sales)
```

***Sales by Product***

```python
product_sales = df.groupby('Product Name')['Sales'].sum()
print(" Total Product :", product_sales)
```

# ***Visualization***


***Bar Chart***

```python
top_10_products = product_sales.nlargest(10)

top_10_products.plot(kind='bar', figsize=(12, 6))
plt.title("Top 10 Sales by Product")
plt.xlabel("Product Names")
plt.ylabel("Sales")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
